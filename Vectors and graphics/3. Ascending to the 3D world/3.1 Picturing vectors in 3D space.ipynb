# Exercise 3.2 --- Mini Project

There are exactly eight 3D vectors whose coordinates are all either +1 or 1. For instance, (1, -1, 1) is one of these. 

Plot all of these eight vectors as points. 

Then figure out how to connect them with line segments using Segment3D objects to form the outline of a cube.

```python
points = [ ]
possibility = [1, -1]

for a in possibility:
    for b in possibility:
        for c in possibility:
            points.append((a,b,c))

print(points)
```

```python
from itertools import combinations

possible_combinations = list(combinations(points, 2))

draw3d(
    *[Segment3D(p[0], p[1]) for p in possible_combinations]
)
```

```python
from math import sqrt

a = (1,1,1) 
b = (1,1,0)

def distance(a,b):
    s = 0
    for j, k in zip(a,b):
        s += (j-k)**2
    return sqrt(s)

distance(a,b)
```

```python
possible_combinations = [p for p in possible_combinations if distance(p[0], p[1]) == 2.0]

draw3d(
    *[Segment3D(p[0], p[1]) for p in possible_combinations]
)
```
